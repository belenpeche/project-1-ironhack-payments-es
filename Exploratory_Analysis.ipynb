```python
# Convert 'created_at' data type to to datetime
cash['created_at'] = pd.to_datetime(cash['created_at'])
```

```python
# Convert all date/time columns in cash dataframe to datetime data type -

date_time_columns = ['created_at', 'updated_at',
       'moderated_at','reimbursement_date',
       'cash_request_received_date', 'money_back_date',
       'send_at', 'reco_creation', 'reco_last_update']

cash[date_time_columns] = cash[date_time_columns].apply(pd.to_datetime, errors = 'coerce')

cash.dtypes
```

```python
# Convert status column to category data type. Converting text data with a
# limited number of unique values to the category type can significantly reduce
# memory usage.
cash['status'] = cash['status'].astype('category')

# Get the different categories present in the status column
status_categories = cash['status'].cat.categories

for category in status_categories:
    print(category)
```

```python
# Create subset of columns by commenting out columns that we don't want to use right now

cash_subset = cash[['id', 'amount', 'status', 'created_at', 'updated_at', 'user_id',
       # 'moderated_at', 'deleted_account_id', 'reimbursement_date',
       # 'cash_request_received_date', 'money_back_date', 'transfer_type',
       'send_at', 'recovery_status', 'reco_creation', 'reco_last_update']].copy()

cash_subset.head(5)
```

```python
# Count total values in money_back_date column. This does not include NaN (Not a Number) entries
count_values_money_back_date = cash['money_back_date'].count()
count_values_money_back_date
```

```python
# Set 'created_at' column as index
cash_copy = cash.copy()
cash_copy.set_index('created_at', inplace=True)

# Group by month and count the number of operations per month. 'M' stands for month end frequency. Other options include W, D, Q, Y (week, day, quarter,)
operations_by_month = cash_copy.resample('M').size()

print(operations_by_month)

# The output shows the last date of each month. In pandas, when you use the
# resample() function to group data by time periods like months, weeks, etc.,
# the default behavior is to label the groups with the end timestamp of each
# period.
```

```python
#Count number of occurrences of each user_id in the cash dataframe. Displayed in descending order.

user_id_counts = cash['user_id'].value_counts()
user_id_counts
```

```python
# Count total number of unique user IDs in the dataframe (see also 'length' of previous ouput)
unique_user_ids = cash['user_id'].nunique()
unique_user_ids
```

```python
# Create month and cohorts columns
cash.sort_values(by='user_id', inplace=True)
cash['month'] = cash['created_at'].dt.to_period('M')
cash['cohorts'] = cash.groupby('user_id')['month'].transform('min')
cash[['cohorts', 'month', 'user_id']].head(10)
```
